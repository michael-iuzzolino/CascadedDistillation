```python
import os
import scipy.io
import glob
import sys
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader
import torchvision
from PIL import Image
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import torch
sys.path.append('..')
from datasets import cub2011_handler
```

```python
root = '/hdd/mliuzzolino/datasets'
glob.glob(f'{root}/*')
```

```python
datasets = cub2011_handler.create_datasets(root)#, split_idxs_root='split_idxs_test_DELETE_ME')
```

```python
nrow = 5
train_loader = DataLoader(datasets['train'], batch_size=nrow**2, shuffle=True)
val_loader = DataLoader(datasets['val'], batch_size=nrow**2, shuffle=False)
test_loader = DataLoader(datasets['test'], batch_size=nrow**2, shuffle=False)
```

```python
loaders = {"train": train_loader, "val": val_loader, "test": test_loader}
for dataset_key, loader in loaders.items():
  for batch_i, (X, y) in enumerate(loader):
    break

  grid_im = make_grid(X, nrow=nrow, normalize=True).permute(1,2,0)
  plt.figure(figsize=(12,12))
  plt.imshow(grid_im)
  title = dataset_key + ' y=['
  for yi in y:
    title += f'{yi},'
  title = title[:-1]
  title += ']'
  plt.title(title)
```
