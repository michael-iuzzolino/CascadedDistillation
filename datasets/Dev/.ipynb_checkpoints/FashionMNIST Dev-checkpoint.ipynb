```python
import os
import sys
sys.path.append('../')
from datasets import fashion_mnist_handler
import torch
from torch.utils.data import DataLoader
import torchvision
from PIL import Image
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
```

```python
dataset_root = '/hdd/mliuzzolino/datasets'
```

```python
datasets = fashion_mnist_handler.create_datasets(dataset_root, val_split=0.1, split_idxs_root='tmp')
```

```python
nrow = 5
train_loader = DataLoader(datasets['train'], batch_size=nrow**2, shuffle=False)
val_loader = DataLoader(datasets['val'], batch_size=nrow**2, shuffle=False)
test_loader = DataLoader(datasets['test'], batch_size=nrow**2, shuffle=False)
```

```python
loaders = {"train": train_loader, "val": val_loader, "test": test_loader}
for dataset_key, loader in loaders.items():
  for batch_i, (X, y) in enumerate(loader):
    break

  grid_im = make_grid(X, nrow=nrow, normalize=True).permute(1,2,0)
  plt.figure(figsize=(8,8))
  plt.imshow(grid_im)
  plt.title(dataset_key)
  plt.axis('off')
```
